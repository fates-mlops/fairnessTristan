# <span style="color:#1E90FF;"> Définitions détaillées des acronymes liés à la transparence en IA </span>

### <a id="IRIT"></a><span style="color:#1E90FF;">IRIT</span>  
**Institut de Recherche en Informatique de Toulouse**

Centre de recherche multidisciplinaire rassemblant des experts en IA, systèmes critiques, analyse de données, et interaction homme-machine. L’IRIT joue un rôle clé dans les réflexions sur l’éthique et la régulation de l’intelligence artificielle.


### <a id="FATES"></a><span style="color:#1E90FF;">FATES</span>  
**Fairness, Accountability, Transparency, Ethics, Security**

Cadre structurant les exigences éthiques et sociotechniques en IA :  
- *Fairness* (équité)  
- *Accountability* (responsabilité)  
- *Transparency* (explicabilité)  
- *Ethics* (éthique)  
- *Security* (sécurité)


### <a id="MLOPS"></a><span style="color:#1E90FF;">MLOps</span>  
**Machine Learning Operations**  

Pratiques de type DevOps appliquées au déploiement, à la maintenance et à la surveillance des modèles de machine learning en production.


### <a id="HAL"></a><span style="color:#1E90FF;">HAL</span>  
**Hyper Articles en Ligne**  

Plateforme française d’archives ouvertes pour publications scientifiques et données de recherche.  
[voir le site HAL](https://hal.archives-ouvertes.fr)


### <a id="ML"></a><span style="color:#1E90FF;">ML</span>  
**Machine Learning (Apprentissage automatique)**  

Ensemble de techniques permettant à des algorithmes d’apprendre à partir des données.


### <a id="COMPAS"></a><span style="color:#1E90FF;">COMPAS</span>  
**Correctional Offender Management Profiling for Alternative Sanctions**  

Algorithme prédictif dans la justice pénale américaine, critiqué pour son manque de transparence et ses biais.

### <a id="AI-Driven"></a><span style="color:#1E90FF;">AI-Driven</span>  
Piloté par l’IA  

Système ou processus où l’intelligence artificielle joue un rôle central dans la prise de décision.

### <a id="ASL"></a><span style="color:#1E90FF;">AI Safety Level (ASL)</span>

Le **AI Safety Level** correspond à une classification des systèmes d’IA selon leur niveau de sécurité, basé sur les risques et impacts potentiels qu’ils peuvent engendrer. 

- **Niveau 1** : Sécurité faible  
  Systèmes d’IA avec des risques limités, dont les conséquences négatives sont peu probables ou peu graves. Généralement utilisés dans des contextes où une erreur a un faible impact.

- **Niveau 2** : Sécurité modérée  
  Systèmes pouvant présenter des risques plus importants, nécessitant des mécanismes de contrôle et de surveillance pour limiter les effets indésirables potentiels.

- **Niveau 3** : Sécurité élevée  
  Systèmes critiques où les défaillances peuvent entraîner des conséquences graves sur les personnes, les biens ou l’environnement. Une attention forte est portée à la robustesse, la fiabilité et la gestion des risques.

- **Niveau 4** : Sécurité très élevée  
  Systèmes avec un potentiel de risque majeur, qui requièrent des garanties strictes, des audits réguliers, et une supervision humaine constante pour assurer la sécurité.

Cette classification aide à adapter les exigences de sécurité aux différents usages et permet d’orienter le développement et la régulation des IA selon leur criticité.


### <a id="API"></a><span style="color:#1E90FF;">API</span>  
**Application Programming Interface**  

Interface permettant la communication entre logiciels ou entre un utilisateur et un système d’IA.


### <a id="GDPR"></a><span style="color:#1E90FF;">GDPR</span>  
**General Data Protection Regulation (RGPD)** 

Règlement européen encadrant la protection des données personnelles.  
[Lien vers europa](https://eur-lex.europa.eu/eli/reg/2016/679/oj)


### <a id="AIA"></a><span style="color:#1E90FF;">AIA</span>  
**Artificial Intelligence Agent**  

Agent autonome doté de capacités d’apprentissage, de décision et d’action.


### <a id="FERM"></a><span style="color:#1E90FF;">FERM</span>  
**Fair Empirical Risk Minimization**  

Méthode visant à équilibrer précision prédictive et équité dans les modèles d’apprentissage.  
[lien d'explication vers IEEE Xplore](https://ieeexplore.ieee.org/abstract/document/9206819)


### <a id="VFAE"></a><span style="color:#1E90FF;">VFAE</span>  
**Variational Fair Autoencoder**  

Autoencodeur conçu pour produire des représentations de données neutres vis-à-vis d’attributs sensibles.


### <a id="EIA"></a><span style="color:#1E90FF;">EIA</span>  
**Ethical Impact Assessment**  

Évaluation des risques éthiques liés à l’usage d’un système d’IA.


### <a id="SHAP"></a><span style="color:#1E90FF;">SHAP</span>  
**SHapley Additive exPlanations**  

Méthode d’explicabilité individuelle inspirée de la théorie des jeux. 


### <a id="LIME"></a><span style="color:#1E90FF;">LIME</span>  
**Local Interpretable Model-agnostic Explanations**  

Méthode qui explique localement une prédiction par un modèle simplifié.  


### <a id="WIT"></a><span style="color:#1E90FF;">WIT</span>  
**What-If Tool**  

Outil visuel pour tester et analyser les comportements des modèles de ML.  


### <a id="HITL"></a><span style="color:#1E90FF;">HITL</span>  
**Human-In-The-Loop**  

Approche dans laquelle l’humain intervient activement dans la boucle de décision de l’IA.


### <a id="HOTL"></a><span style="color:#1E90FF;">HOTL</span>  
**Human-On-The-Loop**  

Modèle où l’humain surveille le fonctionnement d’un système automatisé sans intervenir systématiquement.


### <a id="BIAS"></a><span style="color:#1E90FF;">BIAS</span>  
**Business Impact Analysis**

Analyse des effets et risques d’un système IA sur les processus métiers, notamment les biais.
